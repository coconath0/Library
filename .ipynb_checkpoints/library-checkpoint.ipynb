# TXST Datathon: Library Impact on Student Success (2015-2024)
### Research Question
To what extent does library circulation volume track with student enrollment and degree completion rates at Texas State, and can physical circulation serve as a leading indicator for student engagement in a digital era?
## Problem Statement
As Texas State University (TXST) grows, understanding how students engage with library resources is critical for future resource planning. This project analyzes the relationship between physical book circulation and key university metrics (Enrollment & Completion) to determine if traditional library usage remains a primary driver of academic success.

### Datasets Used
* **TXST Internal Library Reports (FY15-FY24):** Yearly circulation and borrowing statistics.
* **IPEDS Admissions Data (2019-2024):** Official university enrollment totals.
* **IPEDS Completers Data (2019-2024):** Official degree completion totals.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define Library Circulation Data (Based on extracted report values)
lib_data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Circulation': [75018, 69605, 70349, 68784, 59898, 48179, 86065, 61659, 64315, 64090]
}
lib_df = pd.DataFrame(lib_data)

# 2. Load and Filter IPEDS Data for Texas State (UNITID: 228459)
txst_id = 228459

# Enrollment Data
adm_df = pd.read_csv('IPEDS Admissions Data 2019-2024.csv')
enroll_df = adm_df[adm_df['UNITID'] == txst_id].groupby('Year')['Enrolled total'].max().reset_index()
enroll_df.rename(columns={'Enrolled total': 'Enrollment'}, inplace=True)

# Degree Completers Data
comp_df = pd.read_csv('IPEDS Completers Data 2019-2024.csv')
comp_sum = comp_df[comp_df['UNITID'] == txst_id].groupby('Year')['Grand total'].sum().reset_index()
comp_sum.rename(columns={'Grand total': 'Completers'}, inplace=True)

# 3. Final Merge
df = lib_df.merge(enroll_df, on='Year', how='left').merge(comp_sum, on='Year', how='left')

# 4. Calculate Key Metrics
df['Circ_Per_Student'] = df['Circulation'] / df['Enrollment']
print("--- Processed Data ---")
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean, academic style
sns.set_theme(style="whitegrid")

# FIGURE 1: The Divergence (Line Chart)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Circulation Plot
color_circ = '#003366'  # Dark Professional Blue
ax1.set_xlabel('Academic Year', fontsize=12, fontweight='bold')
ax1.set_ylabel('Annual Library Circulation', color=color_circ, fontsize=12, fontweight='bold')
ax1.plot(df['Year'], df['Circulation'], marker='o', color=color_circ, linewidth=4, label='Circulation')
ax1.tick_params(axis='y', labelcolor=color_circ)
ax1.grid(True, linestyle='--', alpha=0.6)

# Enrollment Plot
ax2 = ax1.twinx()
color_enroll = '#990000' # Deep Maroon (fits the TXST theme)
ax2.set_ylabel('Student Enrollment (IPEDS)', color=color_enroll, fontsize=12, fontweight='bold')
ax2.plot(df[df['Enrollment'].notnull()]['Year'], df[df['Enrollment'].notnull()]['Enrollment'], 
         marker='s', linestyle='--', color=color_enroll, linewidth=3, label='Enrollment')
ax2.tick_params(axis='y', labelcolor=color_enroll)

plt.title('Analysis of Library Usage vs. Institutional Growth (2015-2024)', fontsize=15, pad=20)
fig.tight_layout()
plt.savefig('txst_divergence_analysis.png', dpi=300)
plt.show()

# FIGURE 2: Engagement vs. Success (Scatter Plot)
plt.figure(figsize=(10, 6))
# Focus on the years where we have both library and IPEDS data (2019-2024)
clean_subset = df.dropna(subset=['Circulation', 'Completers'])

sns.regplot(data=clean_subset, x='Circulation', y='Completers', 
            scatter_kws={'s':100, 'color':'#501214'}, line_kws={'color':'black', 'ls':'--'})

plt.title('Correlation: Library Circulation vs. Degree Completers', fontsize=14)
plt.xlabel('Annual Circulation Count', fontsize=12)
plt.ylabel('Total Degree Completers', fontsize=12)
plt.savefig('correlation_success.png', dpi=300)
plt.show()
```

```python
# Select only the metrics we want to compare
plot_df = df[['Circulation', 'Enrollment', 'Completers']].dropna()

# Calculate the correlation matrix
corr = plot_df.corr()

# Create the Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap: Library Engagement & Student Success', fontsize=14)
plt.savefig('txst_success_heatmap.png', dpi=300)
plt.show()
```
